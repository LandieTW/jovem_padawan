```python
import os
import pandas as pd
import psycopg2
```

```python
_this_path = os.getcwd()
key_id = pd.read_csv(
    os.path.join(_this_path, "db_entry.csv")
)
```

```python
connection = psycopg2.connect(
    host = key_id.loc[0, 'Entry Door'],
    port = 5432,
    database = 'postgres',
    user = 'postgres',
    password = key_id.loc[0, 'Password'],
)
```

```python
connection.autocommit = True
cursor = connection.cursor()
```

```python
cursor.execute('drop table arquivos')
cursor.execute('create table arquivos (idarquivo INT, nomearquivo VARCHAR(256));')
```

```python
cursor.execute("insert into arquivos (idarquivo, nomearquivo) values (123, 'test.jpg')")
```

```python
cursor.execute('select * from arquivos')
recursive_search = cursor.fetchall()
for rec in recursive_search:
    print(rec)
```

```python
cursor.execute('delete from arquivos')
```

```python
connection.close()
```
