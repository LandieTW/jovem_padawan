```python
import boto3
import pandas as pd
import os
import io
from io import BytesIO
```

```python
_this_path = os.getcwd()
key_id = pd.read_csv(
    os.path.join(_this_path, "rootkey.csv")
)
```

```python
s3 = boto3.resource(
    service_name = 's3',
    region_name = 'sa-east-1',    # SP
    aws_access_key_id = key_id.loc[0, 'Access key ID'],
    aws_secret_access_key = key_id.loc[0, 'Secret access key']
)
```

```python
bucket_name = 'landierandoms3bucket'
folder_name = 'images'
```

```python
for obj in s3.Bucket(bucket_name).objects.filter(Prefix=folder_name):
    if obj.key.endswith(('jpg', 'png')):
        
        filename = obj.key.split('/')[1]
        print(filename)

        body = obj.get()['Body'].read()
        img = io.BytesIO(body)

        img_path = os.path.join(_this_path, "images_cloud")
        os.makedirs(img_path, exist_ok=True)
        
        with open(os.path.join(img_path, filename), 'wb') as f:
            f.write(img.getbuffer())
```
